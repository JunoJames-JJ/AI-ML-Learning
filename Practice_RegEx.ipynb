```python
import re

# Sample text
text = """Contact us at support@example.com or sales@example.org.
You can also call +1-202-555-0173 or 202-555-0199 for more information.
The cat chased the scatter of leaves in the catalog.
He likes dogs, but not hotdogs.
We stood by the riverbank after visiting the bank downtown.
Her username is test_user123 and her backup email is user.name@university.edu.
The meeting is scheduled for 21/09/2025 at 10:30 AM.
Order IDs: A1234, B-5678, and C_90X.
Visit https://example.com or http://test.org for details.
Prices: $5.99, €12.50, and £100 are listed in the catalog.
Welcome!
"""

```

```python
print(re.findall(r'\d', text))
```

```python
print(re.findall(r'\d{3}', text))
```

```python
print(re.findall('[1-5]{3}', text)) # another way by using range [0-9]
```

```python
print(re.findall('[0-2]{3}', text))
```

```python
print(re.findall(r'\D', text))
```

```python
print(re.findall(r'\w', text)) # word characters
```

```python
print(re.findall(r'\W', text)) # non-word
```

```python
print(re.findall(r'\s', text))
```

```python
print(re.findall(r'\S', text))
```

```python
print(re.findall('cat', text))
```
