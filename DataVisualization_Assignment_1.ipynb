   b. Create a box plot to visualize the distribution of ages based on passenger class.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8,6))
# hue = split data into subgroups & color them
sns.boxplot(x='class', y='age', data=titanic, palette='pastel', hue='sex')
plt.title('Distribution of Ages by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

  c. Generate a bar plot showing the count of male and female passengers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

# Count of male and female passengers
counts = titanic['sex'].value_counts()


plt.figure(figsize=(8,6))
sns.barplot(x=counts.index, y=counts.values)
plt.title('Count of Male and Female Passengers')
plt.xlabel('Gender')
plt.ylabel('Number of passengers')
plt.show()
```

2. Using the Iris dataset (iris = sns.load_dataset('iris')):
--------------------------------------------------------------

a. Calculate the mean, median, and standard deviation of the sepal length for each species. Display these summary statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

print("iris data: \n", iris.head())
```

```python
print("iris data: \n", iris.describe())
```

```python
# calculating the mean, median, and std
summary = iris.groupby('species')['sepal_length'].agg(['mean', 'median', 'std'])
print(summary)
```

b. Create a scatter plot for 'petal length' vs. 'petal width' with different colors and sizes for each species. The size of the points should represent the ratio of 'sepal length' to 'sepal width' for each data point.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

# creating the size ratio of  of 'sepal length' to 'sepal width'
iris['size_ratio'] = iris['sepal_length'] / iris['sepal_width']

plt.figure(figsize=(8,6))

sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species', size='size_ratio', palette='pastel', alpha=0.7)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
