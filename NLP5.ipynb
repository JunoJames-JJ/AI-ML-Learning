```python
with open("pg11.txt", "r", encoding="utf-8") as f:
    alice = f.read()

    start_index = alice.find("Alice was beginning")
    end_index = alice.find("conversations?") + len("conversation?")

    sample = alice[start_index:end_index]

print(sample)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(sample)

print("Tokens: ")
for token in doc:
  print(token.text)
```
