```python
sample = alice[start_index:2000]

doc = nlp(sample)

sentences = [sent.text for sent in doc.sents]

print(sentences)
len(sentences)
```

```python
for sent in doc.sents:
  print("sentence:", sent.text)
  print("Start index: ", sent.start)
  print("End index: ", sent.end)
  print("Char positions: ", sent.start_char, "-", sent.end_char)
  print("Noun chunks: ", [chunk.text for chunk in sent.noun_chunks])
  print()
```

```python
# Normalization:
normalized = [token.text.lower() for token in doc
              if token.is_alpha and not token.is_stop]

lemmas = [token.lemma_.lower() for token in doc
          if token.is_alpha and not token.is_stop]

print("After normalization: ", normalized[:30])
print("After lemmitization: ", lemmas[:30])


```

```python
import pandas as pd

rows = []
for token in doc:
  rows.append([token.text, token.pos_, spacy.explain(token.pos_), token.tag_, spacy.explain(token.tag_)])

  df = pd.DataFrame(rows, columns=["Token", "POS", "POS explanation", "Tag", "Tag explanation"])

  df.to_csv("POS_Explanation.csv")
  print(df)
```
